```python
# Simple Linear Regression - Predicting Salary Based on Experience

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Sample Dataset: Years of Experience vs Salary
data = {
    'Experience': [1, 2, 3, 4, 5, 6 ,8,9,10],
    'Salary': [35000, 40000, 45000, 50000, 60000, 65000,75000, 80000, 85000]
}
df = pd.DataFrame(data)
```

```python
#preprocessing
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python

# Splitting features and target
X = df[['Experience']]
y = df['Salary']
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Model Training
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Evaluation
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)
```

```python

# Visualization
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, model.predict(X), color='red', label='Predicted')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```
