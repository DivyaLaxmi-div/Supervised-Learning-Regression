```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
import sklearn.datasets
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.datasets import fetch_california_housing
```

```python
housing = fetch_california_housing()
```

```python
print(housing)
```

```python
df=pd.DataFrame(housing.data,columns=housing.feature_names)
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
#correlation
correlation=df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Greens')
```

```python
df['price']=housing.target
```

```python
#splitting the data and target
x=df.drop(['price'],axis=1)
y=df['price']
```

```python
print(x)
print(y)
```

```python
#splitting
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
print(x_train.shape,x_test.shape)
```

```python
#model training
model=XGBRegressor()
model.fit(x_train,y_train)
```

```python
x_predict=model.predict(x_train)
score1=metrics.r2_score(y_train,x_predict)
score2=metrics.mean_absolute_error(y_train,x_predict)
print('r_Square:',score1)
print('mae:',score2)
```

```python
#visualizing
plt.scatter(y_train,x_predict)
plt.xlabel('Actual prices')
plt.ylabel('predicted prices')
plt.title('Actuall price vs predicted prices')
plt.show()
```

```python
#prediction of text data
x_predict=model.predict(x_test)
score1=metrics.r2_score(y_test,x_predict)
score2=metrics.mean_absolute_error(y_test,x_predict)
print('r_Square:',score1)
print('mae:',score2)
```
